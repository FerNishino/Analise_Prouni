```python
#Quantidade de cursos por UF
data['uf_busca'].value_counts()
```

O Estado SP possui a maior quantidade de cursos

```python
counts = data['uf_busca'].value_counts()
ax = counts.plot(kind='bar', rot=0)
ax.set_title('Quantidade de cursos por Estado')
ax.set_xlabel('Estado')
ax.set_ylabel('Quantidade de cursos')
```

```python
#Qual o curso com mensalidade mais cara?
data[data['mensalidade'] == data['mensalidade'].max()]
```

```python
#Qual o curso mais barato do Brasil?
data[data['mensalidade'] == data['mensalidade'].min()]
```

```python
#Qual o curso mais caro por cada estado?
data.groupby('uf_busca')['mensalidade'].max()
```

```python
#Qual o curso mais barato por cada estado?
data.groupby('uf_busca')['mensalidade'].min()
```

```python
#Mostrar todos os dados dos cursos mais caros de cada UF
data['mensalidade'] = data['mensalidade'].str.replace(',', '.')
data['mensalidade'] = data['mensalidade'].astype(float)
data.loc[data.groupby('uf_busca')['mensalidade'].idxmax()]
```
